```python
from scipy.stats import chi2_contingency
def chi_square_test(feature, Salary):
    
    cross_tab = pd.crosstab(df[feature], df[Salary])
    chi2_stat, p_val, dof, ex = chi2_contingency(cross_tab)
    print(f"Chi-Square Test for {feature} vs {Salary}")
    print(f"Chi-Square Statistic: {chi2_stat}, p-value: {p_val}\n")
```

```python
features = ['Designation', 'JobCity','10board','12board','Specialization','Gender','Degree','CollegeState']
for feature in features:
    chi_square_test(feature, 'Salary')
```
